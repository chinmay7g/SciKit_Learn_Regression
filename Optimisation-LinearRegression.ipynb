```python
#importing

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating data

X = np.linspace(0,1,100)
exp = np.random.choice([2,3])

y = X**exp + np.random.randn(X.shape[0])/10
```

```python
plt.plot(X,y,'.')
```

```python
#sklearn

from sklearn.linear_model import LinearRegression

linReg = LinearRegression(fit_intercept = True)
```

```python
Xt = X.reshape(-1,1)
```

```python
linReg.fit(Xt,y)
```

```python
plt.plot(X,y,'.')
plt.plot(Xt, linReg.predict(Xt), 'r')
```

```python
linReg.intercept_, linReg.coef_[0]
```

```python
linReg.score(Xt,y)
```

# Gradient Descent

```python
beta0 = linReg.intercept_
beta1 = np.linspace(0,1)

MSE = [((y - (beta0 + m * Xt))**2).sum() for m in beta1]
```

```python
plt.plot(beta1, MSE)
```

# Huber Loss

```python
from sklearn.linear_model import HuberRegressor
huber = HuberRegressor(fit_intercept= True, alpha=0)
huber.fit(Xt,y)
```

```python
huber.predict(Xt)
```

```python
huber.score(Xt,y), linReg.score(Xt,y)
```

```python
plt.plot(Xt,y,'.')
plt.plot(Xt, linReg.predict(Xt), 'r-')
plt.plot(Xt, huber.predict(Xt), 'y-')
plt.legend([ '','MSE','Huber'])
```
